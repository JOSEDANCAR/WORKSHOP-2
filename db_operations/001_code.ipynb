Establecimiento de la Conexión

```python
import mysql.connector
from mysql.connector import Error

try:
    # Establecer la conexión
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="root",
        database="spotify_grammy_db"
    )

    if conn.is_connected():
        print("Conexión exitosa a la base de datos")

        # Crear un cursor
        cursor = conn.cursor()

        # Ejecutar una consulta
        cursor.execute("SHOW TABLES;")

        # Imprimir los resultados
        for table in cursor:
            print(table)

except Error as e:
    print(f"Error al conectar a la base de datos: {e}")

finally:
    if 'conn' in locals() and conn.is_connected():
        # Cerrar la conexión
        conn.close()
        print("Conexión a la base de datos cerrada.")

```

OPERACIONES SPOTIFY


```python
import pandas as pd
import mysql.connector
from sqlalchemy import create_engine

# Definir los parámetros de conexión a MySQL
db_user = 'root'
db_password = 'root'
db_host = 'localhost'
db_name = 'spotify_grammy_db'
table_name = 'spotify_dataset'

# Conexión usando SQLAlchemy para realizar la inserción
engine = create_engine(f'mysql+pymysql://{db_user}:{db_password}@{db_host}/{db_name}')

# Cargar el archivo CSV
csv_file_path = '../Data/spotify_dataset.csv'
df = pd.read_csv(csv_file_path)

# Eliminar la columna 'Unnamed: 0' si existe
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Reemplazar NaN con valores predeterminados
df = df.fillna({
    'track_id': 'Unknown',
    'artists': 'Unknown',
    'album_name': 'Unknown',
    'track_name': 'Unknown',
    'popularity': 0,
    'duration_ms': 0,
    'explicit': False,
    'danceability': 0.0,
    'energy': 0.0,
    'key': 0,
    'loudness': 0.0,
    'mode': 0,
    'speechiness': 0.0,
    'acousticness': 0.0,
    'instrumentalness': 0.0,
    'liveness': 0.0,
    'valence': 0.0,
    'tempo': 0.0,
    'time_signature': 0,
    'track_genre': 'Unknown'
})

# Insertar los datos en la tabla de MySQL
try:
    df.to_sql(table_name, engine, if_exists='append', index=False)
    print(f"Datos insertados exitosamente en la tabla '{table_name}' de la base de datos '{db_name}'.")
except Exception as e:
    print(f"Error al insertar los datos: {e}")
```

```python
import mysql.connector
import pandas as pd

# Conexión a MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",  # Usuario root
    password="root",  # Contraseña root
    database="spotify_grammy_db"  # Base de datos
)

# Crear un cursor
cursor = conn.cursor()

# Consulta para obtener las primeras 10 filas de la tabla spotify_dataset
query = "SELECT * FROM spotify_dataset LIMIT 10;"
cursor.execute(query)

# Obtener los datos de la consulta
data = cursor.fetchall()

# Obtener los nombres de las columnas
column_names = [i[0] for i in cursor.description]

# Crear un DataFrame de pandas con los datos obtenidos
df = pd.DataFrame(data, columns=column_names)

# Cerrar la conexión
cursor.close()
conn.close()

# Mostrar los datos
print(df)
```
